```python
# functions practice
def last_char(name):
    return name[-1]

print(last_char("vivek"))   
#last_char(9) # error
```

```python
def odd_even(num):
    if num%2 == 0:
        return "even"
    else:
        return "odd"

print(odd_even(10))           
```

```python
def is_even(num):
    if num%2 == 0: # true
        return True
    else:
        return False
print(is_even(9))            
```
