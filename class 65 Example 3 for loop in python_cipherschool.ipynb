```python
# practice for loop
# ask user name and count each character
# "Vivek Kumar"
# V : 1
# i : 2
# v : 3
# e : 4
# k : 5

name = input("ENTER YOUR NAME")
temp = ""
for i in range(len(name)):
    if name[i] not in temp:
        print(f"{name[i]}: {name.count(name[i])}")
        temp += name[i]
```
